# overview of campaign deployments

```python
%matplotlib inline
import matplotlib.pyplot as plt

# cognac data and tools
import cognac.insitu as insi
```

```python
#campaign = 'microco_leg2.yaml'
campaign = 'lopstech19_leg1.yaml'
```

```python
cp = insi.utils.campaign(campaign)
```

```python
fig, ax, crs = insi.utils.plot_map(cp=cp)
for uname, u in cp.items():
    for d in u:
        if uname=='ctd':
            line=False
        else:
            line=True
        if uname=='source':
            yshift=2
        else:
            yshift=0.
        d.plot_on_map(ax, line=line, label=uname+'_'+d.label, color=u['color'], 
                     yshift=yshift)
```

```python
fig = plt.figure()
ax=fig.add_subplot(111)

y=0
yticks, yticks_labels = [], []
for uname, u in cp.items():
    for d in u:
        ax.plot([d.start.time,d.end.time],[y,y], lw=1, color=u['color'])
    yticks.append(y)
    yticks_labels.append(uname)
    y+=1
    
ax.set_yticks(yticks)
ax.set_yticklabels(yticks_labels)
```
